## Actions

```python
def read_modify_write(in_fname):
    
    modified=False
    if 'bayesfusion' in in_fname:
        return
    elif '.csv' not in in_fname:
        return
    else:
        # Read
        df = pd.read_csv(in_fname, header=None)
        
        # Modify
        for col in df:
            if df[col].nunique() < 2:
                df = df.drop([col], axis=1)
                modified=True
        # Write
        if modified:
            out_fname = in_fname
            df.to_csv(out_fname, index=False, header=False)
        
            msg = """
            Succesful modification of file: {}
            Results written to: {}
            """.format(in_fname, out_fname)
            print(msg)
        
    return
```

```python
def fix_dir(directory):
    ds_fnames = [os.path.join(directory, x) for x in os.listdir(directory)]
    ds_fnames.sort()
    for f in ds_fnames:
        read_modify_write(f)
    return
```

```python
for ds in datasets:
    ds_dir = os.path.join(data_dir, ds)
    fix_dir(ds_dir)
    msg = """
    Done with dataset: {}
    """.format(ds)
    print(msg)
```

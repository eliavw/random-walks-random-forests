```python
ds_name = os.listdir(tidy_dir)
ds_dirs = [os.path.join(tidy_dir, ds) for ds in ds_name]
```

```python
def summarize_dataset(ds_name, ds_dir, pos_filters=(), neg_filters=()):
    fnames = [fn for fn in os.listdir(ds_dir)]
    
    pos_filters += (ds_name,)
    
    for f in pos_filters:
        fnames = [fn for fn in fnames
                  if f in fn]
        
    for f in neg_filters:
        fnames = [fn for fn in fnames
                  if f not in fn]
        
    msg = """
    Loading files: {}
    """.format(fnames)
    print(msg)
    
    dfs = [pd.read_csv(os.path.join(ds_dir, fn), header=None) for fn in fnames]
    
    nb_instances, nb_attributes = pd.concat(dfs).shape
    
    return ds_name, nb_instances, nb_attributes
```

```python
def build_dataset_summary(ds_names):
    ds_dirs = [os.path.join(tidy_dir, ds) for ds in ds_name]
    
    it = zip(ds_names, ds_dirs)
    
    
    res = [summarize_dataset(n, d, pos_filters=('F00',), neg_filters=('bayesfusion',))
          for n,d in it]
    
    return res
```

```python
res = build_dataset_summary(ds_name)
```

```python
df = pd.DataFrame(res)
df.columns = ('Dataset name', 'Number of instances', 'Number of attributes')
df.sort_values('Dataset name')
```
